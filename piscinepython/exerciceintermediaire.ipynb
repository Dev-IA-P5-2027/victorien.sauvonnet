# **exercice 2**

```python


def Fonction(x):
    result=int(2*x**3+x-5)
    return result

def poly(fonction,borninf:int,bornsup:int,touslesnbpas:int)->dict:
    liste_x=[]
    liste_fx=[]
    for i in range(borninf,bornsup+1,touslesnbpas):
                liste_fx.append(fonction(i))
                liste_x.append(i)
    return dict(zip(liste_x,liste_fx))
      


print(poly(Fonction,0,10,2))
```

# **exercice 4** 

```python
import re
def compterMots(chaines_de_characteres:str):
    liste_de_mots=re.split("[ ,.?]",chaines_de_characteres)
    doublon=[]
    dict_mots={}

    for i in liste_de_mots :

        if i not in doublon :
            dict_mots[i]=1
            doublon.append(i)
        else :

            dict_mots[i]= dict_mots[i]+1

    return dict_mots

compterMots("bounjour les gens comment, ca va les gens comment ca va pas. ca va bien ?" )    
```
