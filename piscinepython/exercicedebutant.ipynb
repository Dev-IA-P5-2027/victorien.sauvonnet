# **exercice 2**

```python
def  volume_cone(rayon:float,hauteur:float)-> float :
    from math import pi
    Vcone=(1/3)*pi*(rayon**2)*hauteur
    return Vcone
print(volume_cone(1,3))
```

# **exercice 3**

```python
def IMC(poids_kg:float,taille_M:float)->float :
    if poids_kg<=0 : return "poids invalide"
    if taille_M<=0 : return "taille invalide"
    result= round(poids_kg/(taille_M**2),2)
    return result
print(IMC(-2,1.7))
```

# **exercice 4**

```python
def conv_imperial_to_metrique(vitesse_en_milesheures):
    kilometre_heure=vitesse_en_milesheures*1.609
    metre_par_seconde=(kilometre_heure*1000)/3600
    return round(kilometre_heure,2), round(metre_par_seconde,2)
print(conv_imperial_to_metrique(120))
```

# **exercice 6**

```python
import random
score=int()
for i in range(1,6) :
    premier_nb=random.randint(1,10)
    deuxieme_nb=random.randint(1,10)
    proposition = int(input(f"quel est le resultat de la multiplication entre {premier_nb} et {deuxieme_nb}:"))
    if proposition==deuxieme_nb*premier_nb :
        score+=1
    else : 
        score-=1
print(f"votre score est de {score}")
```

# **exercice 9**

```python

notes=[]
while True :
    saisie=input("entre une note :")
    if saisie!="" : notes.append(int(saisie))

    if saisie=="":
        break

print(notes)
print(f"l'élève a {len(notes)} notes")
print(f"ça meilleur note est {max(notes)}")
print(f"ça moins bonne note est {min(notes)}")
print(f"ca moyenne est de {sum(notes)/len(notes)}")
```
